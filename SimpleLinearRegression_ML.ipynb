** Regression - Machine learning model to predict continous target values **

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X1 = np.array([1,2,3,4.3,5,8,7,9])              # feature
y1 = np.array([1,4,9,16,25,36,49,64])           # target
# not a categorical feature -> its not like that same target for multiple features -> so that prediction would be continous in nature from -infinity to +infinity --> dealing with regression problem
```

```python
# simple linear regression ---> having just single feature

plt.scatter(X1,y1)
plt.xlabel('X1')           # almost linear line
plt.ylabel('y1')
plt.show()
```

```python
# for best fit line --> we have two methods
# 1.> Ordinary-least square method for Linear Regression -->
# 2.> Gradient descent method for Linear Regression --->
```

** Ordinary-least square method for Linear Regression **

```python
# The ordinary least squares (OLS) algorithm is a method for estimating the parameters of a linear regression model. The OLS algorithm aims to find the values of the linear regression model's parameters (i.e., the coefficients) that minimize the sum of the squared residuals
# y = mx+c -> equation of best fit line --> m = slope and c = intercept of best fit line of the data
#yh = w0 + w1*x1

#find slope of best fit line-->
w1 = ((X1 - X1.mean()) * (y1 - y1.mean())).sum()/((X1 - X1.mean())**2).sum()
print("Slope of best fit line" , w1)

#find intercept of best fit line-->
w0 = y1.mean() - w1*X1.mean()
print("Intercept of best fit line" , w0)
```

```python
yh = w0 + w1 * X1
```

```python
# plot the best fit line -->
plt.scatter(X1,y1)
plt.plot(X1,yh,color='red')
plt.xlabel('X1')
plt.ylabel('y1')         # data is distributed non-linearly but the relation btw x and y are linear for best fit line
plt.show()
```

```python
# new predicted data point -->
X1_new = 6.5
y1_new = w0 + w1*X1_new
print("Predicted value for X1_new" , y1_new)
```

```python
# evaluate error rate -->
# evaluate average error
mae = abs(y1 - yh).mean()
print("Mean Absolute Error" , mae)
```

** Gradient Descent for Linear Regression **

```python
# Gradient Descent is an optimization algorithm for finding a local minimum of a differentiable function. Gradient descent in machine learning is simply used to find the values of a function's parameters (coefficients) that minimize a cost function as far as possible.
# Gradient descent is an optimization algorithm used to train machine learning models by minimizing errors between predicted and actual results.
```

```python
# taking different data -->
X1 = np.array([3, 4.3, 5, 8, 7, 9])         #feature
y1 = np.array([4.2, 6, 7.1, 8, 9, 12])      #target
```

```python
# equation for bfl --> yh = w0 + w1x1
# step1 ->  find out values of w0 & w1 such that difference between y1 and yh is minimum

# don't know exact value for w0 and w1 so take any value --->
w0 = 2
w1 = 2

# after considering any value calculate yh ->
yh = w0 + w1*X1
print(yh) # this yh is not a best fit line so for this gd method is used
```

```python
plt.scatter(X1,y1)
plt.plot(X1,yh,color='red')
plt.xlabel('X1')
plt.ylabel('y1')                  # not a bfl -> gd method is used
plt.show()
```

```python
# Gradient descent --> used for multiple linear regression , logistic regression and neural networks then gradient descent will be useful
# steps -> calculate error value
error = ((y1 - yh)**2).mean()
print(error)
```

```python
# steps -> reduce the error rate --> achive global minima point where error is minimum
# calculate gradient for w0 and w1 -->
grw0 = -2*(y1 - yh).mean()
grw1 = -2*(X1 *(y1 - yh)).mean()
print(grw0)
print(grw1)
```

```python
# learning rate -> control it
lr = 0.01
w0 = w0 - lr * grw0
w1 = w1 - lr * grw1
print(w0, w1)
```

```python
yh = w0 + w1 * X1
error = ((y1 - yh)**2).mean()
print(error)                 # error value scaled from 45 to 2
```

```python
# all steps in single loop -->
for i in range(0,100):
    yh = w0 + w1 * X1

    grw0 = -2*(y1 - yh).mean()
    grw1 = -2*(X1 *(y1 - yh)).mean()

    w0 = w0 - lr * grw0
    w1 = w1 - lr * grw1

    error = ((y1 - yh)**2).mean()
    print(i,error)
```

```python
plt.scatter(X1,y1)
plt.plot(X1,yh,color='red')
plt.xlabel('X1')
plt.ylabel('y1')
plt.show()
```
